import jam cam videos that have been uploaded to s3 into python environment (and code to download onto local drives or onto virtual environments)

```python
import boto3
import io
import base64
from IPython.display import HTML
import cv2
import numpy as np
```

```python
# session required because of our access designation
s3_session = boto3.Session(profile_name='dssg')
print(type(s3_session))
```

```python
# INPUTS: change key name according to which video you want
bucket_name = 'air-pollution-uk'
key_name = 'raw/video_data/06508/2019-06-03_11.45.mp4'
localfile_name = 'c:/Users/joh3146/Desktop/video001.mp4' # or other path
```

```python
s3_resource = s3_session.resource('s3')
my_bucket = s3_resource.Bucket(bucket_name)
```

download file to local machine

```python
# download the file to your local machine
my_bucket.download_file(key_name, localfile_name)
```

upload the file to be viewed

```python
# view downloaded file embedded in jupyter notebook
video = io.open(localfile_name, 'r+b').read()
encoded = base64.b64encode(video)
HTML(data='''<video alt="test" controls>
                <source src="data:video/mp4;base64,{0}" type="video/mp4" />
             </video>'''.format(encoded.decode('ascii')))
```

upload file as a numpy array

```python
cap = cv2.VideoCapture(localfile_name)
frameCount = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
frameWidth = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frameHeight = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

buf = np.empty((frameCount, frameHeight, frameWidth, 3), np.dtype('uint8'))

fc = 0
ret = True

while (fc < frameCount  and ret):
    ret, buf[fc] = cap.read()
    fc += 1

cap.release()

cv2.namedWindow('frame 10')
cv2.imshow('frame 10', buf[9])

cv2.waitKey(0)
```
