```python
import numpy as np 
import pandas as pd
```

### Initial analytics on YOLO results 

```python
csv_path = "C:\\Users\\Caroline Wang\\OneDrive\\DSSG\\air_pollution_estimation\\data\\sample_yolo_output\\sample_yolo_output.csv"
yolo_df = pd.read_csv(csv_path)
yolo_df.head(15)
```

##### Get counts of objects detected per frame

```python
obj_counts_frame = yolo_df.groupby(["frame_id", "obj_classification"]).size().reset_index(name = 'obj_count')
print(obj_counts_frame.head())
obj_counts_frame=obj_counts_frame.pivot(index='frame_id', columns='obj_classification', values='obj_count').fillna(value = 0)
print(obj_counts_frame.head())
```

##### Get summary statistics of num and types of objects detected over all frames 

```python
sums = obj_counts_frame.aggregate(func = "sum")

summary = obj_counts_frame.describe()
summary = summary.drop(["count", "25%", "75%"], axis = 0)
summary.loc['sum'] = sums
summary.head(10)
```
